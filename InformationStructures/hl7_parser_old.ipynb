```python
json_dict = dict()

for i, line in enumerate(obs):
    msh = line.split("|")[0]
    json_dict[msh] = {}
    line = line.replace("\n", "")
    for ii, field in enumerate(line.split("|")):
        #construct key with MESSAGE HEADER.i.ii
        key = line.split("|")[0]+"."+str(i)+"."+str(ii)
        if len(field)==0 or field=="" or field==" ":
            continue
        json_dict[msh][key] = {}
        
        for iii, segment in enumerate(field.split("^")):
            key_segment = key+"."+str(iii)
            if len(field.split("^")) > 1:
                if len(segment)==0 or segment=="" or segment==" ":
                    continue
                json_dict[msh][key][key_segment] = segment
            else:
                if len(segment)==0 or segment=="" or segment==" ":
                    continue
                json_dict[msh][key] = field
```
