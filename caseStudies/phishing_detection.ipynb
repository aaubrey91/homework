```python
import pandas as pd
import matplotlib.pyplot
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("dataset.csv")
```

```python
#Ensure its read correctly.
df.head()
```

```python
#don't need the index column..
df.drop(['index'], inplace=True, axis=1)
```

```python
# Reference: 
# http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
x_train, x_test, y_train, y_test = train_test_split(df.loc[:, df.columns != "Result"], 
                                                    df['Result'], 
                                                    test_size=0.33, 
                                                    random_state=42)
```

```python
clfs = [LogisticRegression(), BernoulliNB(), RandomForestClassifier(), AdaBoostClassifier(), MLPClassifier(), svm.SVC(), KNeighborsClassifier()]
title = ['LogisticRegression', 'BernoulliNB', 'RandomForestClassifier', 'AdaBoostClassifier','MLPClassifier', 'SVM Classifier', 'KNN']

for t,clf in zip(title, clfs):
    model = clf
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    print(t,"Performance\n=========================")
    print(model.classes_)
    print(confusion_matrix(y_test, pred))
    print("Accurracy:", accuracy_score(y_test, pred)) 
    print("")
```
